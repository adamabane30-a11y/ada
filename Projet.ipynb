```python

catalogue = []
ventes = []

def enregistrer_dans_fichier(texte):
    fichier = open("historique.txt", "a" ,encoding="UTF-8")
    fichier.write(texte + "\n")
    fichier.close()

def afficher_menu():
    print("\n      GESTION POINT DE VENTE      ")
    print("1. Ajouter un produit au catalogue")
    print("2. Afficher le catalogue")
    print("3. Enregistrer une vente")
    print("4. Rechercher un produit (Bonus)")
    print("5. Afficher les statistiques")
    print("6. Quitter le programme")

def main():
    enregistrer_dans_fichier("\n--- DÉBUT DE SESSION ---")

    while True:
        afficher_menu()
        choix = input("Entrez votre choix (1-6) : ").strip()
        if choix == "1":
            ajouter_produit(catalogue)
        elif choix == "2":
            afficher_catalogue(catalogue)
        elif choix == "3":
            enregistrer_vente()
        elif choix == "4":
            rechercher_produit()
        elif choix == "5":
            afficher_statistiques()
        elif choix == "6":
            print("👋 Au revoir ! Merci d'avoir utilisé notre application.")
            enregistrer_dans_fichier("--- FIN DE SESSION ---\n")
            break
        else:
            print("❌ Choix invalide ! Saisissez un nombre entre 1 et 6.")

def ajouter_produit(catalogue):
    nom = input("Nom du produit : ").strip()

    for produit in catalogue:
        if produit[0].lower() == nom.lower():
            print(f"❌ Le produit '{produit[0]}' existe déjà.")
            return

    categories_autorisees = ["smartphone", "ordinateur", "accessoire"]

    while True:
        print("\n📢 Catégories disponibles : Smartphone, Ordinateur, Accessoire")
        categorie = input("Catégorie du produit : ").strip()

        if categorie.lower() in categories_autorisees:
            categorie = categorie.capitalize()
            break
        else:
            print("❌ Erreur : Saisissez une catégorie valide.")

    try:
        montant = float(input("Prix du produit (FCFA) : "))
        if montant <= 0:
            print("❌ Le prix doit être supérieur à 0.")
            return

        catalogue.append([nom, montant, categorie])
        print(f"✅ Produit '{nom}' ({categorie}) ajouté.")

        enregistrer_dans_fichier(f"[CATALOGUE] Ajout du produit : {nom} | Catégorie: {categorie} | Prix: {montant} FCFA")

    except ValueError:
        print("❌ Erreur : Saisissez un prix valide.")

def afficher_catalogue(catalogue):
    print("\n=== CATALOGUE DES PRODUITS ===")
    if len(catalogue) == 0:
        print("Le catalogue est vide.")
    else:
        for produit in catalogue:
            print(f"- {produit[0]} [{produit[2]}] : {produit[1]} FCFA")

def appliquer_reduction(montant):
    taux = 0.05
    if montant > 20000:
        taux = 0.05
    else:
        taux = 0
    reduction = montant * taux
    print(f"Taux de réduction appliqué : {taux * 100}%")
    return reduction

def enregistrer_vente():
    if len(catalogue) == 0:
        print("❌ Aucun produit enregistré.")
        return

    produit_vendu = input("Produit vendu : ").strip()
    indice = -1
    for i in range(len(catalogue)):
        if catalogue[i][0].lower() == produit_vendu.lower():
            indice = i
            break

    if indice == -1:
        print("❌ Produit introuvable.")
        return

    try:
        quantite = int(input("Quantité : "))
        if quantite <= 0:
            print("❌ La quantité doit être supérieure à 0.")
            return

        total = catalogue[indice][1] * quantite
        reduction = appliquer_reduction(total)
        montant_final = total - reduction

        ventes.append({
            "produit": catalogue[indice][0],
            "quantite": quantite,
            "montant": montant_final
        })

        print(f"✅ Vente enregistrée ! À payer : {montant_final} FCFA")

        enregistrer_dans_fichier(f"[VENTE] {catalogue[indice][0]} - Quantité: {quantite}, Montant Final: {montant_final} FCFA (Réduction: {reduction} FCFA)")

    except ValueError:
        print("❌ Erreur : Saisissez une quantité entière.")

def rechercher_produit():
    nom_recherche = input("Nom du produit à rechercher : ").strip()
    for produit in catalogue:
        if produit[0].lower() == nom_recherche.lower():
            print(f"🔍 Trouvé : {produit[0]} [Catégorie: {produit[2]}] - Prix : {produit[1]} FCFA")
            enregistrer_dans_fichier(f"[RECHERCHE] Produit recherché : {nom_recherche} - Trouvé")
            return
    print("❌ Ce produit n'est pas dans le catalogue.")

    enregistrer_dans_fichier(f"[RECHERCHE] Produit recherché : {nom_recherche} - Non trouvé")

def afficher_statistiques():
    print("\n=== STATISTIQUES DU POINT DE VENTE ===")
    if len(ventes) == 0:
        print("❌ Aucune vente enregistrée.")
        return

    chiffre_affaires = 0
    total_articles = 0
    dict_produits = {}

    for v in ventes:
        chiffre_affaires += v["montant"]
        total_articles += v["quantite"]
        dict_produits[v["produit"]] = dict_produits.get(v["produit"], 0) + v["quantite"]

    enregistrer_dans_fichier(f"[STATISTIQUES] CA: {chiffre_affaires} FCFA, Articles vendus: {total_articles}")

    produit_star = max(dict_produits, key=dict_produits.get)

    print(f"💰 Chiffre d'affaires : {chiffre_affaires:.2f} FCFA")
    print(f"🧾 Nombre de ventes  : {len(ventes)}")
    print(f"📦 Articles vendus   : {total_articles}")
    print(f"🛒 Panier moyen      : {chiffre_affaires / len(ventes):.2f} FCFA")
    print(f"⭐ Produit star      : {produit_star} ({dict_produits[produit_star]} unités)")

if __name__ == "__main__":
    main()
```
